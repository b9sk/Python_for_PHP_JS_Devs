#### **–ò—Ç–µ—Ä–∞—Ü–∏—è –ø–æ —Å–ª–æ–≤–∞—Ä—è–º:**

```python

    print(f"–ö–ª—é—á: {key}")

    print(f"–ó–Ω–∞—á–µ–Ω–∏–µ: {value}")
```

**–ó–ê–î–ê–ù–ò–ï:** –ü–µ—Ä–µ–±–µ—Ä–∏ —Å–ª–æ–≤–∞—Ä—å `student` –∏ –≤—ã–≤–µ–¥–∏ –≤—Å–µ –µ–≥–æ –∫–ª—é—á–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏—è. üëá

````python
student = {"name": "–û–ª–µ–≥", "course": "Python", "grade": "A"}

print("–ö–ª—é—á–∏:")
for key in student.keys():
    print(key)

print("\n–ó–Ω–∞—á–µ–Ω–∏—è:")
for value in student.values():
    print(value)

print("\n–ü–∞—Ä—ã –∫–ª—é—á-–∑–Ω–∞—á–µ–Ω–∏–µ:")
for key, value in student.items():
    print(f"{key}: {value}")
````
