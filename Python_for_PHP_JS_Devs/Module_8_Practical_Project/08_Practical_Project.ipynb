### üß† **2. –°–æ–∑–¥–∞–µ–º –ø–µ—Ä–≤–æ–µ API: "Hello, Flask!"**

–°–æ–∑–¥–∞–¥–∏–º –ø—Ä–æ—Å—Ç–µ–π—à–µ–µ Flask-–ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ. –í —ç—Ç–æ–π –∂–µ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ —Å–æ–∑–¥–∞–π —Ñ–∞–π–ª `app.py` —Å–æ —Å–ª–µ–¥—É—é—â–∏–º —Å–æ–¥–µ—Ä–∂–∏–º—ã–º:

```python
# app.py
from flask import Flask, jsonify, request

app = Flask(__name__)

# –ü—Ä–æ—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö (–∏–º–∏—Ç–∞—Ü–∏—è –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö)
books = [
    {"id": 1, "title": "–í–æ–π–Ω–∞ –∏ –º–∏—Ä", "author": "–õ–µ–≤ –¢–æ–ª—Å—Ç–æ–π"},
    {"id": 2, "title": "–ü—Ä–µ—Å—Ç—É–ø–ª–µ–Ω–∏–µ –∏ –Ω–∞–∫–∞–∑–∞–Ω–∏–µ", "author": "–§–µ–¥–æ—Ä –î–æ—Å—Ç–æ–µ–≤—Å–∫–∏–π"}
]
next_id = 3

@app.route('/')
def home():
    return "–ü—Ä–∏–≤–µ—Ç –∏–∑ Flask! –≠—Ç–æ –Ω–∞—à API –¥–ª—è –∫–Ω–∏–≥. üìö"

if __name__ == '__main__':
    app.run(debug=True) # debug=True –¥–ª—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–æ–π –ø–µ—Ä–µ–∑–∞–≥—Ä—É–∑–∫–∏ –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏—è—Ö
```

**–ó–ê–î–ê–ù–ò–ï:**

### ‚öôÔ∏è **3. –†–µ–∞–ª–∏–∑–∞—Ü–∏—è CRUD-–æ–ø–µ—Ä–∞—Ü–∏–π –¥–ª—è –∫–Ω–∏–≥**

–¢–µ–ø–µ—Ä—å –¥–æ–±–∞–≤–∏–º —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –Ω–∞—à–∏–º —Å–ø–∏—Å–∫–æ–º –∫–Ω–∏–≥. –ú—ã —Ä–µ–∞–ª–∏–∑—É–µ–º:
- **GET /books**: –ü–æ–ª—É—á–∏—Ç—å –≤—Å–µ –∫–Ω–∏–≥–∏.
- **GET /books/<id>**: –ü–æ–ª—É—á–∏—Ç—å –∫–Ω–∏–≥—É –ø–æ ID.
- **POST /books**: –î–æ–±–∞–≤–∏—Ç—å –Ω–æ–≤—É—é –∫–Ω–∏–≥—É.
- **PUT /books/<id>**: –û–±–Ω–æ–≤–∏—Ç—å —Å—É—â–µ—Å—Ç–≤—É—é—â—É—é –∫–Ω–∏–≥—É.
- **DELETE /books/<id>**: –£–¥–∞–ª–∏—Ç—å –∫–Ω–∏–≥—É.

**–ó–ê–î–ê–ù–ò–ï:** –î–æ–±–∞–≤—å —Å–ª–µ–¥—É—é—â–∏–π –∫–æ–¥ –≤ `app.py` (–≤–º–µ—Å—Ç–æ –∏–ª–∏ –ø–æ—Å–ª–µ —Å—É—â–µ—Å—Ç–≤—É—é—â–µ–≥–æ `home` —Ä–æ—É—Ç–∞). –û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ `request` –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –∏–∑ –∑–∞–ø—Ä–æ—Å–∞ –∏ `jsonify` –¥–ª—è –æ—Ç–ø—Ä–∞–≤–∫–∏ JSON-–æ—Ç–≤–µ—Ç–æ–≤. üëá

```python
# app.py (–¥–æ–ø–æ–ª–Ω–µ–Ω–∏—è)

# –ü–æ–ª—É—á–∏—Ç—å –≤—Å–µ –∫–Ω–∏–≥–∏
@app.route('/books', methods=['GET'])
def get_books():
    return jsonify(books)

# –ü–æ–ª—É—á–∏—Ç—å –∫–Ω–∏–≥—É –ø–æ ID
@app.route('/books/<int:book_id>', methods=['GET'])
def get_book(book_id):
    book = next((b for b in books if b['id'] == book_id), None)
    if book:
        return jsonify(book)
    return jsonify({"message": "–ö–Ω–∏–≥–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞"}), 404

# –î–æ–±–∞–≤–∏—Ç—å –Ω–æ–≤—É—é –∫–Ω–∏–≥—É
@app.route('/books', methods=['POST'])
def add_book():
    global next_id
    new_book = {
        "id": next_id,
        "title": request.json['title'],
        "author": request.json['author']
    }
    books.append(new_book)
    next_id += 1
    return jsonify(new_book), 201

# –û–±–Ω–æ–≤–∏—Ç—å –∫–Ω–∏–≥—É
@app.route('/books/<int:book_id>', methods=['PUT'])
def update_book(book_id):
    book = next((b for b in books if b['id'] == book_id), None)
    if book:
        book['title'] = request.json.get('title', book['title'])
        book['author'] = request.json.get('author', book['author'])
        return jsonify(book)
    return jsonify({"message": "–ö–Ω–∏–≥–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞"}), 404

# –£–¥–∞–ª–∏—Ç—å –∫–Ω–∏–≥—É
@app.route('/books/<int:book_id>', methods=['DELETE'])
def delete_book(book_id):
    global books
    initial_len = len(books)
    books = [b for b in books if b['id'] != book_id]
    if len(books) < initial_len:
        return jsonify({"message": "–ö–Ω–∏–≥–∞ —É–¥–∞–ª–µ–Ω–∞"}), 200
    return jsonify({"message": "–ö–Ω–∏–≥–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞"}), 404
```
