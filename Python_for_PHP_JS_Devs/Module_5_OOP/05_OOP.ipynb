**–ó–ê–î–ê–ù–ò–ï:** –°–æ–∑–¥–∞–π –∫–ª–∞—Å—Å `Student`, –∫–æ—Ç–æ—Ä—ã–π –Ω–∞—Å–ª–µ–¥—É–µ—Ç –æ—Ç `Person` (–∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è). –î–æ–±–∞–≤—å –≤ `Student` –Ω–æ–≤–æ–µ —Å–≤–æ–π—Å—Ç–≤–æ `student_id` –∏ –º–µ—Ç–æ–¥ `study()`, –∫–æ—Ç–æ—Ä—ã–π –≤—ã–≤–æ–¥–∏—Ç `[–ò–ú–Ø] —É—á–∏—Ç—Å—è.`. üëá

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"–ü—Ä–∏–≤–µ—Ç, –º–µ–Ω—è –∑–æ–≤—É—Ç {self.name} –∏ –º–Ω–µ {self.age} –ª–µ—Ç. üëã")

class Student(Person): # Student –Ω–∞—Å–ª–µ–¥—É–µ—Ç –æ—Ç Person
    def __init__(self, name, age, student_id):
        super().__init__(name, age) # –í—ã–∑—ã–≤–∞–µ–º –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–æ–≥–æ –∫–ª–∞—Å—Å–∞
        self.student_id = student_id

    def study(self):
        print(f"{self.name} (ID: {self.student_id}) —É—á–∏—Ç—Å—è. üìö")

my_student = Student("–°–≤–µ—Ç–∞", 20, "S12345")
my_student.introduce() # –ú–µ—Ç–æ–¥ –æ—Ç Person
my_student.study() # –ú–µ—Ç–æ–¥ –æ—Ç Student
````
