### üé© **3. –î–µ–∫–æ—Ä–∞—Ç–æ—Ä—ã: "–§—É–Ω–∫—Ü–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –æ–±–æ—Ä–∞—á–∏–≤–∞—é—Ç —Ñ—É–Ω–∫—Ü–∏–∏"**

–î–µ–∫–æ—Ä–∞—Ç–æ—Ä—ã ‚Äî —ç—Ç–æ –º–æ—â–Ω—ã–π —Å–ø–æ—Å–æ–± –∏–∑–º–µ–Ω—è—Ç—å –∏–ª–∏ —Ä–∞—Å—à–∏—Ä—è—Ç—å –ø–æ–≤–µ–¥–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π –∏–ª–∏ –º–µ—Ç–æ–¥–æ–≤ –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏—è –∏—Ö –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –∫–æ–¥–∞. –≠—Ç–æ –∫–∞–∫ "–æ–±–µ—Ä—Ç–∫–∞" –≤–æ–∫—Ä—É–≥ —Ñ—É–Ω–∫—Ü–∏–∏. –í –¥—Ä—É–≥–∏—Ö —è–∑—ã–∫–∞—Ö –µ—Å—Ç—å –ø–æ—Ö–æ–∂–∏–µ –∫–æ–Ω—Ü–µ–ø—Ü–∏–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∞–Ω–Ω–æ—Ç–∞—Ü–∏–∏ –≤ Java –∏–ª–∏ –∞—Ç—Ä–∏–±—É—Ç—ã –≤ C#, –Ω–æ –≤ Python —ç—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–π.

–ü—Ä–æ—Å—Ç–æ–π –¥–µ–∫–æ—Ä–∞—Ç–æ—Ä –≤—ã–≥–ª—è–¥–∏—Ç —Ç–∞–∫:

```python
def my_decorator(func):
    def wrapper():
        print("–ß—Ç–æ-—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –î–û –≤—ã–∑–æ–≤–∞ —Ñ—É–Ω–∫—Ü–∏–∏.")
        func() # –í—ã–∑—ã–≤–∞–µ–º –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é

    return wrapper

@my_decorator # –≠—Ç–æ —Å–∏–Ω—Ç–∞–∫—Å–∏—á–µ—Å–∫–∏–π —Å–∞—Ö–∞—Ä –¥–ª—è say_hello = my_decorator(say_hello)
```

**–í—ã–≤–æ–¥:**
```

–ü—Ä–∏–≤–µ—Ç, –º–∏—Ä!
–ß—Ç–æ-—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ü–û–°–õ–ï –≤—ã–∑–æ–≤–∞ —Ñ—É–Ω–∫—Ü–∏–∏.
```

–î–µ–∫–æ—Ä–∞—Ç–æ—Ä—ã —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è –ª–æ–≥–∏—Ä–æ–≤–∞–Ω–∏—è, –ø—Ä–æ–≤–µ—Ä–∫–∏ –ø—Ä–∞–≤ –¥–æ—Å—Ç—É–ø–∞, –∫—ç—à–∏—Ä–æ–≤–∞–Ω–∏—è –∏ —Ç.–¥.

**–ó–ê–î–ê–ù–ò–ï:** –°–æ–∑–¥–∞–π –¥–µ–∫–æ—Ä–∞—Ç–æ—Ä `timer`, –∫–æ—Ç–æ—Ä—ã–π –∏–∑–º–µ—Ä—è–µ—Ç –≤—Ä–µ–º—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –≤—ã–≤–æ–¥–∏—Ç –µ–≥–æ. üëá

````python
import time

def timer(func):
    def wrapper():
        start_time = time.time()
        func()
        end_time = time.time()
        print(f"–§—É–Ω–∫—Ü–∏—è '{func.__name__}' –≤—ã–ø–æ–ª–Ω–∏–ª–∞—Å—å –∑–∞ {end_time - start_time:.4f} —Å–µ–∫—É–Ω–¥.")
    return wrapper

@timer
def long_running_task():
    print("–ù–∞—á–∏–Ω–∞–µ–º –¥–æ–ª–≥—É—é –∑–∞–¥–∞—á—É...")
    time.sleep(1) # –ò–º–∏—Ç–∏—Ä—É–µ–º –¥–æ–ª–≥—É—é —Ä–∞–±–æ—Ç—É
    print("–ó–∞–¥–∞—á–∞ –∑–∞–≤–µ—Ä—à–µ–Ω–∞.")

long_running_task()
````
