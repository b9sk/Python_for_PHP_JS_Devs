**–ó–ê–î–ê–ù–ò–ï:** –°–æ–∑–¥–∞–π —Ñ—É–Ω–∫—Ü–∏—é `power(base, exponent=2)`, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤–æ–¥–∏—Ç `base` –≤ —Å—Ç–µ–ø–µ–Ω—å `exponent`. –í—ã–∑–æ–≤–∏ –µ—ë —Å–Ω–∞—á–∞–ª–∞ –±–µ–∑ `exponent`, –ø–æ—Ç–æ–º —Å –Ω–∏–º. üëá

````python
def power(base, exponent=2):
    return base ** exponent

print(f"2 –≤ —Å—Ç–µ–ø–µ–Ω–∏ 2 (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é): {power(2)}")
print(f"2 –≤ —Å—Ç–µ–ø–µ–Ω–∏ 3: {power(2, 3)}")
print(f"3 –≤ —Å—Ç–µ–ø–µ–Ω–∏ 4 (–∏–º–µ–Ω–æ–≤–∞–Ω–Ω—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç): {power(base=3, exponent=4)}")
````

### üèπ **3. `lambda`-—Ñ—É–Ω–∫—Ü–∏–∏: "–ë—ã—Å—Ç—Ä—ã–µ –∞–Ω–æ–Ω–∏–º–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏"**

–ö–∞–∫ –∏ –≤ JS –µ—Å—Ç—å —Å—Ç—Ä–µ–ª–æ—á–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ (`=>`) –∏–ª–∏ –∞–Ω–æ–Ω–∏–º–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ PHP, –≤ Python –µ—Å—Ç—å `lambda`-—Ñ—É–Ω–∫—Ü–∏–∏. –≠—Ç–æ –Ω–µ–±–æ–ª—å—à–∏–µ –∞–Ω–æ–Ω–∏–º–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–≥—É—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–Ω–æ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ.

**–ó–ê–î–ê–ù–ò–ï:** –°–æ–∑–¥–∞–π `lambda`-—Ñ—É–Ω–∫—Ü–∏—é `is_even`, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —á–∏—Å–ª–æ —á–µ—Ç–Ω—ã–º. üëá
